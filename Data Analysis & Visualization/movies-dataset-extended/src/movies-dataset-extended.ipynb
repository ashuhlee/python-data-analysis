#### **Question 8:** Plotting your data

- Plot the distribution of `Popularity Ranking` using any suitable type of plot that you prefer.
- Add a title to the graphic and the percentage that represents each category.
- Can you show the legend of the plot? Consult the bibliography or Google search to find more information on this, as it is not covered in the lecture. 

```python
import matplotlib.pyplot as plt

# count the frequency of each popularity ranking and sort from highest to lowest
pop_count = df['popularity_ranking'].value_counts().sort_index(ascending=False)

# add custom color palette for pie chart as a list
palette = ['#FFDD96', '#89f5f2', '#afadfd', '#F1ABFD', '#FFADA1', '#85A2FF', '#444973']
bg_color = '#373b58'

# set size of the chart
plt.figure(figsize=(14,10), facecolor=bg_color)

# create a pie chart with percentages
plt.pie(pop_count,
        labels=pop_count.index, # use ranking values as index
        startangle=90, # start the first slice at 90deg
        autopct='%1.1f%%', # display percentages with 1 decimal place
        colors=palette, # apply custom colour palette
        center=(0,0), # center the chart
        textprops={'fontsize': 10, 'fontweight': 'bold', 'color': 'white'}) # text customization (font weight, size and colour)

# add title and customize the style
plt.title('Distribution of Popularity Ranking', fontweight='bold', color=palette[2])

# add a legend showing ranking labels and position the legend at the top right
legend = plt.legend(title='Popularity Rankings',
           title_fontsize=10,
           labels=[f'{rank}'for rank in pop_count.index],
           loc='upper right',
           facecolor=bg_color)

legend.get_title().set_color('white')
legend.get_title().set_fontweight('bold')

# display the chart
plt.show()
```
